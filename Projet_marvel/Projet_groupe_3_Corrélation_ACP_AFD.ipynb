```python
# Extraire la variable 'is_alive' et centrer-réduire la base de données
colonnes_marvel = [a for a in marvel.columns if a != 'is_alive']
data = marvel[colonnes_marvel]
data -= data.mean()
data /= data.std()
data.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 15)
cp = pca.fit_transform(data)
cp.shape
```

```python
## ACP
from sklearn.decomposition import PCA
pca = PCA(n_components = 15)
cp = pca.fit_transform(data)

## Variance expliquée par les 15 premières composantes

from matplotlib import pyplot as plt
plt.figure(figsize=(10, 4))
plt.bar(range(pca.n_components_), pca.explained_variance_ratio_)
for i, var in enumerate(pca.explained_variance_ratio_):
    plt.text(i, var + 0.005, f'{var:.0%}', ha='center')
```

```python
# Courbe cumulative
pca2 = PCA()
cp2 = pca2.fit_transform(data)
plt.figure(figsize=(10, 4))
plt.plot(np.cumsum(pca2.explained_variance_ratio_), '-o')
for i, var in enumerate(np.cumsum(pca2.explained_variance_ratio_)):
    plt.text(i, var + 0.015, f'{var:.0%}', ha='center')
    
## on n 'a définitivement pas une bonne représentation'
```

```python
sum(pca.explained_variance_ratio_)
```

```python
# On voit ainsi que les deux premières composantes principales ne décrivent que 7% de la variance expliquée. les 15 premieres composantes n'expliquent 33,18% de la variance expliquée/ On a donc pas une bonne représentation.  
```

```python
# Nuage de points par classe
target = marvel['is_alive']
plt.plot(cp2[target == 1, 0], cp[target == 1, 1], 'o', label='Vivants', alpha=0.5)
plt.plot(cp2[target == 0, 0], cp[target == 0, 1], 'o', label='Décédés', alpha=0.5)
plt.legend();
```

```python
# Sur ce nuage de points, on remarque que les individus décédés et ceux vivants sont superposés: on ne peut les dissocier. Les deux premières composantes ne permettent donc pas de prédire la variable 'is_alive'. On pouvait s'en douter en ayant que 7% de la variance expliquée.
```

```python
# Nuage de points par cos2
numerateur = np.sum(cp2[:, :2] ** 2, axis=1)
normes2 = np.sum(cp2 ** 2, axis=1)
cosines2 = numerateur / normes2
plt.scatter(cp2[:, 0], cp2[:, 1], c=cosines2, cmap='BrBG', vmin=0, vmax=1,edgecolor='k', linewidth=0.3)
plt.colorbar();
```

```python
# La représentation est mauvaise pour la plupart des points (cos2 <= 0.4). ce qui se justifie avec uniquement 7% de variance expliquée
```

```python
# La qualité de représentation d'une variable par rapport à un certain nombre de composantes principales peut être mesurée par la proportion de la variance totale expliquée par ces composantes.
#Ainsi, la qualité de représentation d'une variable par rapport à ces composantes principales est meilleure si une grande proportion de la variance totale est expliquée.
```

```python
def cercle_correlation(axe1=0, axe2=1):
    racine_lambda_k = np.sqrt(pca.explained_variance_[None, [axe1, axe2]])
    u_kj = pca2.components_[[axe1, axe2]].T
    corrs = racine_lambda_k * u_kj
    plt.plot(corrs[:, 0], corrs[:, 1], 'o')
    theta = np.linspace(0, 2*np.pi, 360)
    plt.plot(np.cos(theta), np.sin(theta))
    plt.gca().axis('equal')
    for name, (corr1, corr2) in zip(pca.feature_names_in_, corrs):
        plt.text(corr1, corr2, name, fontsize=10)
cercle_correlation()
```

```python
# Les variables les mieux expliquées par le premier plan factoriel sont : Sex_Female Characters, Eye_nan, Hair_nan, ID_nan et Sex_Male Characters
# Les variables ID_nan, Hair_nan et Eye_nan sont proches les unes des autres: on peut dire qu'elles sont fortement correlées positivement entre elles.
# Sexe_Female characters et Sexe_Male characters sont symétriquement opposés: elles sont donc fortement correlées négativement.
```

```python
# Pour conclure, on voit ici que l'ACP n'a pas été performante; impossible de réduire la dimension des données. L'explication natuelle vient du fait que nous venons d'effectuer une ACP sur des variables binaires.
# Car de base, l'ACP a été pensé pour des variables plutôt quantitatives. Elle marche donc moins bien sur des variables binaires.
# Le mieux serait de faire une analyse en composantes multiples (ACM) qui est beaucoup plus appropriée pour traiter des variables qualitatives ou binaires.
```

```python
#1.3 AFD
```

```python
## AFD linéaire avec is_alive comme variable de référence
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
fd = lda.fit_transform(data, target)
print(fd.shape)
```

```python
# Comme il y a 2 classes, l'AFD engendre 1 axes discriminant
```

```python
## Projection sur l'axe factoriel
from matplotlib import pyplot as plt
from scipy.stats import gaussian_kde
f0 = gaussian_kde(fd[target == 0, 0])
f1 = gaussian_kde(fd[target == 1, 0])
x = np.linspace(fd.min() - 1, fd.max() + 1, 100)
plt.plot(x, f0(x), label='Décédés')
plt.plot(x, f1(x), label='Vivants')
plt.legend();
```

```python
# Non, on a pas une bonne séparation des données. On voit bien que les deux courbes sont superposées.
```

```python
## Accuracy et matrice de confusion
```

```python
preds = lda.predict(data)
accuracy = (preds == target).mean()
print(f'Accuracy : {accuracy:.1%}')
```

```python
## Une accuracy de 77% ici signifie que l'AFD réalisée prédit correctement 77% de la variabilité totale des données.
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay 
cm = confusion_matrix(target, preds) 
print(cm)
ConfusionMatrixDisplay(cm, display_labels=[ 'Décédés','Survivants']).plot();
```

```python
## La matrice de confusion obtenue n'est pas très satisfaisant : Bien que l'AFD ait bien prédit les survivants, elle a aussi prédit presque tous les décédés comme des survivants. 
## Une option pour rendre l'AFD un peu plus utile serait d'utiliser des techniques de validation croisée : entrainer le modèle sur un échantillon d'apprentissage pui l'évaluer sur un echantillon test.
```
